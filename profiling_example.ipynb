### Profile functions in IPython

**profiling**: getting a detailed report of the execution time, *function by function*
+ Every profiler brings its own overhead that can bias the profiling results (*probe effect*)

First, import necessary files

*Note: you need to install [rasterio](https://anaconda.org/conda-forge/rasterio) and [finao](https://anaconda.org/conda-forge/fiona) first*

```python
from runforesttest import *
```

#### Run a sample function

```python
testit_oldschool(zonefilename,datafilename)
```

#### Time one statement with timeit
+ %timeit: evaluate the time taken by one statement
    + optional: add -r and/or -n to specify number of evaluations
+ %%timeit: evaluate the time taken by a whole cell

```python
%timeit testit_oldschool(zonefilename,datafilename)
```

#### Profile code func by func with prun
+ %prun: profile one statement
+ %%prun: profile the whole cell

To understand the statistics：
+ *tottime*: represents how long the interpreter stays in a given function, excluding the time spent in calls to subfunctions.
+ *cumtime*: similar to *tottime*, but includes the time spent in calls to subfunctions.

Options in prun
+ *-s* : sort the report by a particular column
+ *-q* : quell the pager output
+ *-l* : limit the number of lines displayed or to filter the results by function name
+ *-T* : save the report in a text file

```python
%prun testit_oldschool(zonefilename,datafilename)
```

#### Profile code line by line with lprun
*Note: you need to install [line_profiler](https://anaconda.org/conda-forge/line_profiler) and load line_profile module*

Let's save the lprun report to a file and read it
+ The functions to profile need to be explicitly specified with -f

```python
%load_ext line_profiler
```

```python
%lprun -T lprofoutput -f testit_oldschool testit_oldschool(zonefilename,datafilename)
print(open('lprofoutput', 'r').read())
```

#### Profile the memory usage with mprun
*Note: you need to install [memory_profiler](https://anaconda.org/conda-forge/memory_profiler) and load line_profile module*

To understand the statistics:
+ *increment*: allows us to spot those places in the code where large amounts of memory are allocated

```python
%load_ext memory_profiler
```

```python
%mprun -T mprofoutput -f testit_oldschool testit_oldschool(zonefilename,datafilename)
print(open('mprofoutput', 'r').read())
```
